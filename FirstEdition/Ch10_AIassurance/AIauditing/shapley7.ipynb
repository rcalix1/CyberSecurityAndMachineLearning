```python


tokenizer_name = "distilbert-base-uncased"
model_name     = "distilbert-base-uncased-finetuned-sst-2-english"

tokenizer = transformers.DistilBertTokenizerFast.from_pretrained(tokenizer_name)
model = transformers.DistilBertForSequenceClassification.from_pretrained(model_name).cpu()

pipe = TextClassificationPipeline(model=model, tokenizer=tokenizer, return_all_scores=True)

def score_and_visualize(text):
  prediction = pipe([text])
  print(prediction[0])
  explainer = shap.Explainer(pipe)
  shap_values = explainer([text])
  shap.plots.text(shap_values)

score_and_visualize('I am very happy to have finally finished writing this book '
                    'I have been preparing this material for years '
                    'and I am know really glad to see the finished product')


```
