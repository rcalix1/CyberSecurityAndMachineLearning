```python

import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python

kl_loss = nn.KLDivLoss(reduction="batchmean")
```

```python

input = F.log_softmax(  torch.randn(9, 8, requires_grad=True), dim=1)
input.shape
```

```python

target = F.softmax(torch.rand(9, 8), dim=1)
target.shape
```

```python

target
```

```python


numpy_data = target.numpy()[:,2]


plt.hist(numpy_data, bins=10)
plt.title('Distribution of Tensor Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python

output = kl_loss(input, target)
output
```

```python

tensor2 = torch.rand(9, 8) * 3 + 2
tensor2
```

```python

numpy_data = target2.numpy()[:,2]


plt.hist(numpy_data, bins=10)
plt.title('Distribution of Tensor Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python

target2 = F.softmax(tensor2, dim=1)
target2.shape
```

```python

output = kl_loss(input, target2)
output
```
