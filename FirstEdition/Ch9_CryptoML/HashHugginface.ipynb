```python

m = MD5.new()

string_value = b'buy the stock at $2500 a share'

m.update( string_value )

print( m.digest()     )
print( m.hexdigest()  )
```

```python

def hash_fn( token ):
    m = MD5.new()
    m.update( token )
    return m.hexdigest() 

    
```

```python

passwords = ["egg", "school", "house", "horse"]

for word in passwords:
    res = hash_fn( word.encode() )
    print( word, "," ,res )
    
```

```python

import hashlib
from transformers import AutoTokenizer

# Choose a tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def text_to_hash(text):
    # Tokenize the text
    encoded_text = tokenizer(text, return_tensors="pt")

    # Convert the tokenized text to a string
    token_string = str(encoded_text)

    # Create a hash object
    hash_object = hashlib.sha256(token_string.encode())

    # Get the hexadecimal representation of the hash
    hex_dig = hash_object.hexdigest()

    return hex_dig

# Example usage
text = "This is an example text to be hashed."
hash_value = text_to_hash(text)
print(hash_value)
```
